```python
data = pd.read_csv("all_data.csv")
data.shape
```

```python
Cheddar = 0
pj = 0
alf = 0
data["Modifier"].head()
mods = data['Modifier'][68]
```

```python
#favorite cheeses

for i in range(274010):
    if(data["Modifier"][i]) == 'Cheddar':
        Cheddar+=1
    if(data["Modifier"][i]) == 'Pepper Jack':
        pj+=1
    if(data["Modifier"][i]) == 'Alfredo':
        alf+=1
```

```python
print(Cheddar)
print(pj)
print(alf)
cheeses = [Cheddar, pj, alf]
```

```python
fig, ax = plt.subplots()

x = ["Cheddar", "Pepper Jack", "Alfredo"]
y = [Cheddar, pj, alf]


ax.barh(x, y)

ax.set_ylabel('Orders')
ax.set_title('Favorite Cheeses')

plt.show()
```

```python
gc = pp =br=ba=hm = 0



for i in range(274010):
    if(data["Modifier"][i]) == 'Grilled Chicken':
        gc+=1
    if(data["Modifier"][i]) == 'Pulled Pork':
        pp+=1
    if(data["Modifier"][i]) == 'Brisket':
        br+=1
    if(data["Modifier"][i]) == 'Bacon':
        ba+=1
    if(data["Modifier"][i]) == 'Ham':
        hm+=1

x = ['Grilled Chicken', 'Pulled Pork', 'Brisket', 'Bacon', 'Ham']
y = [gc, pp, br, ba, hm]

sortedi = np.argsort(y)
print(sortedi)
xsort = [x[i] for i in sortedi]
ysort = [y[i] for i in sortedi]




```

```python
fig, ax = plt.subplots()


ax.barh(xsort, ysort)

ax.set_xlabel('Orders')
ax.set_title('Choose your meats')   

plt.show()
```
